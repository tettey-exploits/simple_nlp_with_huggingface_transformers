## NLP with Hugging Face Transformer Library

```python
from transformers import pipeline
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from transformers import Trainer, TrainingArguments
```

### Sentiment Analysis

```python
sent_classifier = pipeline("sentiment-analysis")
sent_response = sent_classifier("I have been waiting on a hugging face course my whole life.") # get a response from the sentiment classifier
print(sent_response)
```

#### With Tokenization Explicitly Done

```python
model_name = "distilbert-base-uncased-finetuned-sst-2-english"
model = AutoModelForSequenceClassification.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

sent_classifier_token = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)
sent_response_token = sent_classifier("I have been waiting on a hugging face course my whole life.") # get a response from the sentiment classifier
print(sent_response_token)
```

```python
# How tokenization works
sequence = "I am going home!"
res = tokenizer(sequence)
print(res)
tokens = tokenizer.tokenize(sequence)
print(tokens)
ids = tokenizer.convert_tokens_to_ids(tokens)
print(ids)
decode = tokenizer.decode(ids)
print(decode)
```

### Text Generation

```python
generator = pipeline("text-generation", model="distilgpt2")
gen_response = generator("SMSGAS is a revolutionary tool that allows customers",
                         max_length=30,
                         num_return_sequences=2)
print(gen_response)
```

### Zero-Shot Classification

```python
zero_shot_classifier = pipeline("zero-shot-classification")
zero_shot_response = zero_shot_classifier("SMSGAS is going to revolutionarize delivery services in Ghana",
                                          candidate_labels=["education", "politics", "military", "business"])
print(zero_shot_response)
```
